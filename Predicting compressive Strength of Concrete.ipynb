# ANN 

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
regressor=Sequential()
```

```python
x.shape
```

```python
regressor.add(Dense(units = 8,activation = "relu", init ="random_uniform"))#input layer
```

```python
regressor.add(Dense(units = 16,activation = "relu", init ="random_uniform"))#hidden layer-1
```

```python
regressor.add(Dense(units = 24,activation = "relu", init ="random_uniform"))#hidden layer-2
```

```python
regressor.add(Dense(units = 1,init ="random_uniform"))#output layer
```

```python
regressor.compile(optimizer = "adam",loss = "mean_squared_error",metrics = ["mse"])
```

```python
regressor.fit(x_train,y_train, epochs = 10000,batch_size = 32)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
yp1=regressor.predict(np.array([[475,0,0,228,0,932,594,180]]))
```

```python
yp1
```
